```python
#Filtering the data : low-pass, high-pass and notch filters
for b in new_BLOCS:
    raw[b].filter(0.5, 120., fir_design="firwin") # This method applies a bandpass filter to the data in "raw".
                                           # Here, we keep only frequencies between 0.5 and 150Hz, the rest will be attenuated.
    
    raw[b].notch_filter([60, 120], filter_length='auto',phase='zero', fir_design="firwin") # The notch filter will only
                                            # remove certain frequencies. Since in North America the power lines oscillates
                                            # at 60Hz, we will remove this frequency and it's first harmonic, 120Hz.
```

```python
ica = mne.preprocessing.ICA(n_components=20, random_state=0).fit(raw['b2'], decim=3, picks = EEG_chs) # We compute an ICA object from the data.
# The n_components argument defines the number of components that we wish to obtain.
# It is also possible to state that 0<n_components<1, in which case it will define the proportion of variance that 
# we want to be explained by the N components.
```

```python
# First, we will start by looking for the component(s) that captured cardiac activity.

ecg_epochs = create_ecg_epochs(raw['b2'], ch_name='ECG') # This function will detect cardiac events (heartbeats) on the ECG signal
                                                      # and store their timestamps in "ecg_epochs".

ecg_inds, ecg_scores = ica.find_bads_ecg(ecg_epochs, ch_name='ECG', threshold=0.30) # Now, this function correlates, at the timestamps
                                                      # specified by "ecg_epochs", the activity on EEG057 (our ECG channel) with
                                                      # each ICA component. The component with the greatest correlation is most
                                                      # likely the one we are looking for !

ica.plot_scores(ecg_scores, ecg_inds); # Let's plot the correlations scores for each ICA components. 
                                      # "ecg_inds" contains the indices of components identified as cardiac activity.

ica.plot_properties(ecg_epochs, picks=ecg_inds, psd_args={'fmax': 35.}, # Plot the topography and variance of the components in
                    image_args={'sigma': 1.});
```

```python
ica.plot_sources(raw['b2'])
```

```python
# Now, let's do the same for ocular blinks. Note that two things have changed :
# - The functions used to detect artifacts are now called "eog" instead of "ecg"
# - The channel to which we correlate our components is now EEG058, the EOG channel
eog_epochs = create_eog_epochs(raw['b2'], ch_name='E127')
eog_inds, eog_scores = ica.find_bads_eog(eog_epochs, ch_name='E127', threshold = 2.3)
ica.plot_scores(eog_scores, eog_inds);
ica.plot_properties(eog_epochs, picks=eog_inds, psd_args={'fmax': 35.})

                    #image_args={'sigma': 1.});
```

```python
#Add information in events to specify if trial with or without pareidolia

def reformat_events_blocs (events):    
    if events[4][2] == 1:
        events[:, 2] = np.where(events[:, 2]==1, 0, events[:, 2])
        events[:, 2] = np.where(events[:, 2]==2, 1, events[:, 2])
        events[:, 2] = np.where(events[:, 2]==3, 2, events[:, 2])
        events[:, 2] = np.where(events[:, 2]==4, 3, events[:, 2])
        events[:, 2] = np.where(events[:, 2]==5, 4, events[:, 2])
    if events[4][2] == 5:
        events[:, 2] = np.where(events[:, 2]==1, 0, events[:, 2])
        events[:, 2] = np.where(events[:, 2]==2, 1, events[:, 2])
        events[:, 2] = np.where(events[:, 2]==3, 2, events[:, 2])
        events[:, 2] = np.where(events[:, 2]==4, 4, events[:, 2])
        events[:, 2] = np.where(events[:, 2]==5, 3, events[:, 2])
    return events
```

```python
events_baseline_b1 = mne.find_events(raw['resting_b1'], stim_channel= 'STI 014')
events_b2 = mne.find_events(raw['b2'], stim_channel= 'STI 014')
events_b3 = mne.find_events(raw['b3'], stim_channel= 'STI 014')
events_b4 = mne.find_events(raw['b4'], stim_channel= 'STI 014')
events_b2_format = reformat_events_blocs(events_b2)
events_b3_format = reformat_events_blocs(events_b3)
events_b4_format = reformat_events_blocs(events_b4)
```

```python
# EPOCHS: 
# Gray : 0.5 to 2.5 --> 0.5 sec of gray and 1.5 sec of cross
# Image_on : 0.5 to 2.5 --> beginning of the image


tmin, tmax = 0, 2.5 # Here we define the amount of time we want to keep before (tmin) and after (tmax) the event.
                       # Thus, we will split our signal into 0.7s chunks.
    
baseline = (0, 0.5) # When epoching, we often want to use some kind of baseline correction. Here, we define that our baseline
                       # starts at the beginning of the epoch and goes until the stimulus occurence (t=0)
    
#reject = {'eeg': 40e-6} # The Epochs() function allows to automatically reject events in which one or more channels exceeds a
                        # threshold value. We can set different values for different types of channels (mag, grad, eeg etc...).
                        # This is useful because even a meticulous pre-processing may leave some noisy segments in the data, and
                        # this method only rejects the contaminated trials while keeping the others intact.
            
picks = mne.pick_types(raw['b2'].info, eeg=True)
                       #eeg=False, eog=False) # We choose to analyse only the magnetometers ('mag')

epochs = mne.Epochs(raw['b2'], events=events_b2_format, event_id=event_id_short, tmin=tmin,
                    tmax=tmax, baseline=baseline, reject=None)
```
